```python
games_after_1990 = df[(df['Release date'] >= '1990-01-01') & (df['Release date'] <= '1995-01-01')]

df_sorted = games_after_1990.sort_values(by='Release date')
cantidad = df_sorted.count()
print(cantidad)
df_sorted.head(10)

```

```python
df = df_sorted.reset_index(drop=True)
df
```
