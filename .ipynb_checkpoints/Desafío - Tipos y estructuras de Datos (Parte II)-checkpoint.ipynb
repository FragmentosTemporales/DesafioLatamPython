```python
import pandas as pd
```

```python
# 1.- Creamos un DataFrame con el archivo entregado

df = pd.read_csv('resumen_resultados_astronautas.csv', delimiter=',', encoding='utf-8')
df.head(10)
```

```python
df.info()
```

```python
# 2.- Creamos la columna con la fórmula entregada

df['IMC'] = df['peso'] / ((df['altura'] / 100) ** 2)
```

```python
df.head()
```

```python
# 3.- Creamos la columna con los promedios de las evaluaciones

df['promedio_evaluaciones'] = (df['evaluacion_fisica'] + df['evaluacion_psicologica'] + df['evaluacion_adaptativa'] +  df['evaluacion_medica']) / 4
```

```python
df.head()
```

```python
# 3.- Filtramos según IMC
%%time
df_filtrado = df[(df['IMC'] >= 18.5) & (df['IMC'] <= 24.9)]
```

```python
df_filtrado.head()
```

```python
# 4.- Filtramos según promedio de evaluaciones mayor o igual a 87

df_filtrado_promedio = df_filtrado[df_filtrado['promedio_evaluaciones'] >= 87]
df_filtrado_promedio.head()
```

```python
# Probamos el filtro utilizando la función query

df_filtrado.query("promedio_evaluaciones >= 87")
```

```python
# Reseteamos el index con .reset_index()

df_filtrado_promedio = df_filtrado_promedio.reset_index(drop=True)
df_filtrado_promedio.head()
```

```python
# 6.- modifica el valor de la columna 'califica'

df_filtrado_promedio['califica'] = 'SI'
```

```python
df_filtrado_promedio
```

```python
df_filtrado_promedio.info()
```

```python
# 7.- Exportamos el dataframe resultante limitando la cantidad de numeros después de la coma (.)

df_filtrado_promedio.to_csv(
    "astronautas_calificados.csv",
    sep=',',
    index=False,
    float_format='%.2f'
)
```
